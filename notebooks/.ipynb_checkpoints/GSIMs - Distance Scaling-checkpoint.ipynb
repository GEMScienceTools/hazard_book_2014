```python
import os
import sys
import numpy as np

HOME = os.environ["HOME"]
sys.path.append(HOME+"/Projects/oq-hazardlib")

from pyproj import Geod
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm

from openquake.hazardlib.gsim.akkar_bommer_2010 import AkkarBommer2010 
from openquake.hazardlib.gsim.chiou_youngs_2008 import ChiouYoungs2008
from openquake.hazardlib.gsim.cauzzi_faccioli_2008 import CauzziFaccioli2008
from openquake.hazardlib.gsim.akkar_2014 import AkkarEtAlRjb2014

from openquake.hazardlib import const
from openquake.hazardlib.source.rupture import Rupture
from openquake.hazardlib.geo.surface.simple_fault import SimpleFaultSurface
from openquake.hazardlib.geo import Point, Mesh, Line
from openquake.hazardlib.scalerel.wc1994 import WC1994
from openquake.hazardlib.gsim.base import SitesContext, RuptureContext, DistancesContext 
from openquake.hazardlib.imt import PGA
from openquake.hazardlib.site import Site, SiteCollection
from openquake.hazardlib.geo.geodetic import distance
```

## Fix the parameters

```python
# Parameters describing the rectangular rupture
lon1, lat1 = (10.0, 45.0)
aspect_ratio = 2.
azimuth = 90.0
mag = 7.0
dip = 45

# Magnitude-scaling relationship 
mscr = WC1994()

# Ground motion prediction equation and intensity measure type
imt = PGA()
gsim_list = []
gsim_list.append(AkkarBommer2010())
gsim_list.append(ChiouYoungs2008())
gsim_list.append(AkkarEtAlRjb2014())
#gsim_list.append(CauzziFaccioli2008())
```

## Create the rupture

```python
# Compute the area of the rupture
area = mscr.get_median_area(mag = mag, rake = 0.0)
length = (area * aspect_ratio)**0.5

# Find the second extreme of the fault trace
g = Geod(ellps='WGS84')
lon2, lat2, backaz = g.fwd(lon1, lat1, azimuth, length*1e3)

# Create the rupture surface 
rupture_surface = SimpleFaultSurface.from_fault_data(
    fault_trace = Line([Point(lon1, lat1), Point(lon2, lat2)]), 
    upper_seismogenic_depth = 0.0, 
    lower_seismogenic_depth = 15.0, 
    dip = dip, 
    mesh_spacing = 10.0)
```

## Fix the site

```python
# Fixing the sites
lons = [10.05, 10.25]
lats = [45.3, 45.3]

# Creating the site collection
sit = []
for lon, lat in zip(lons, lats):
    sit.append(
        Site(location = Point(lon,lat), 
             vs30 = 760, 
             vs30measured = 1, 
             z1pt0 = 100, 
             z2pt5 = 1
             )
        )
sites = SiteCollection(sit)

# Setting parameters
stddev_types = [const.StdDev.TOTAL]
sctx = SitesContext()
rctx = RuptureContext()
dctx = DistancesContext()

# Site parameters
sctx.vs30 = 760.0 * np.ones_like(lons)
sctx.z1pt0 = 100.0 * np.ones_like(lons)
sctx.vs30measured = np.ones_like(lons)
```

```python
# Creating the mesh
mesh = Mesh(np.array(lons), np.array(lats), np.zeros_like(np.array(lons)))

# Compute rupture-sites distance
dctx.rjb = rupture_surface.get_joyner_boore_distance(mesh)
dctx.rhypo = rupture_surface.get_middle_point()
dctx.rx = rupture_surface.get_rx_distance(mesh)

closest_points = rupture_surface.get_closest_points(mesh)
dctx.rrup = distance(closest_points.lons, closest_points.lats, closest_points.depths,
                     mesh.lons, mesh.lats, zeros_like(mesh.lons))
    
# Rupture parameters
rctx.mag = mag
rctx.ztor = rupture_surface.get_top_edge_depth()
rctx.dip = dip
```
